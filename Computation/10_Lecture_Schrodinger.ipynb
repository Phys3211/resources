# Quantum Mechanics

Quantum mechanics describes the behavior of particles at microscopic scales, where classical mechanics fails. The fundamental principle is that the state of a quantum system is described by a wavefunction $\Psi(\vec{r},t)$, which contains all the information about the system and evolves according to Schrödinger’s equation.

## Schrödinger's Equation
Schrödinger's equation describes non-relativistic particles subject to a potential $V(\vec{r},t)$. The formal time-dependent equation takes the form $$i\hbar\frac{\partial \Psi(\vec{r},t)}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\vec{r},t)\right)\Psi(\vec{r},t),$$
where $\hbar$ is Planck's constant and $m$ is the mass of the non-relativistic particle.

We can express the Schrödinger's equation in operator language as $$i\hbar\frac{\partial \Psi}{\partial t} = \hat{H}\Psi,$$
where $\hat{H}$ is the Hamiltonian operator, $$\hat{H} = \frac{\hat{p}^2}{2m} + V(r),$$
where $\hat{p}$ is the momentum operator, $-i\hbar\nabla$. Operators, such as $\hat{H}$, $\hat{p}$, and $\hat{x}$ represent measurable quantities (called *observables*). The eigenvalues of the operators correspond to possible measurement outcomes. For instances, the Hamiltonian eigenvalues correspond to the possible energy states of a particle, $\hat{H}\Psi = E\Psi$.

The time-dependent Schrödinger's equation governs the quantum dynamics of a system, describing how a quantum state evolves over time. The *Born interpretation* states the $|\Psi(r,t)|^2$ is the probability density of finding a particle as a given location.

## Time dependent Schrödinger equation

We want to describe an electron wavefunction by a wavepacket $\Psi (x,y)$ that is a function of position $x$ and time $t$. For simplicity, we restrict our electron to 1D. We assume that the electron is initially localized around $x_0$, and model this by a Gaussian multiplying a plane wave:
$$\Psi(x,t=0)\equiv\psi_0(x)=\exp{\left[-\frac{1}{2}\left(\frac{x-x_0}{\sigma _0}\right)^2\right ]} e^{ik_0x}$$

This wave function does not correspond to an electron with a well defined momentum. However, if the width of the Gaussian $\sigma _0$ is made very large, the electron gets spread over a sufficiently large region of space and can be considered as a plane wave with momentum $k_0$ and a slowly varying amplitude.

The behavior of this wave packet as a function of time is described by the 1D time-dependent Schröedinger equation:
$$i\frac{\partial \Psi}{\partial t}=\hat{H}\Psi(x,t).$$ 

$\hat{H}$ is the Hamiltonian operator:

$$\hat{H}=-\frac{\partial ^2}{\partial x^2}+V(x),$$ 

where $V(x)$ is a time-independent potential, and we chose energy units such that $\hbar=1$ and $2m=1$, which is very close to atomic units (we can always translate them back to other unit systems). 

Schrödinger’s equation is obviously a PDE, and we can use generalizations of the techniques learned in previous sections to solve it. The main observation is that **this** time we have to deal with complex numbers since the function $\Psi (x,t)$ has real and imaginary parts:

$$\psi (x,t) = R(x,t)+iI(x,t).$$ 

In what follows, we will present an alternative method that makes the quantum mechanical nature of this problem more transparent.

## The time-evolution operator


The Scrödinger equation above can be integrated in a formal sense to obtain: 
$$\psi(x,t)=U(t)\psi(x,t=0)$$ 
where we have the time-evolution operator $U(t)=\exp({-iHt})$. More generally this gives
$$\psi(t+\Delta t)= e^{-iH\Delta t}\psi(t).$$
Likewise, the inverse of the time-evolution operator moves the wave function back in time:
$$\psi(t-\Delta t)=e^{iH\Delta t}\psi(t),$$ 
where we have use the property $$U^{-1}(t)=U(-t).$$ Although it would be nice to have an algorithm based on the direct application of $U$, however, such a method has been shown to be unstable. Hence, we apply the following relation:

$$\psi(t+\Delta t)=\psi(t-\Delta t)+\left[e^{-iH\Delta t}-e^{iH\Delta t}\right]\psi(t).$$ 

Now, in the case at hand for the above Hamiltonian, the derivatives with respect to $x$ can be
approximated by, as before,

$$\begin{aligned}
\frac{\partial \psi}{\partial t}
&\sim& \frac{\psi(x,t+\Delta t)-\psi(x, t)}{\Delta t}, \\
\frac{\partial ^2 \psi}{\partial x^2} &\sim& \frac{\psi(x+\Delta  %
x,t)+\psi(x-\Delta x,t)-2\psi(x,t)}{(\Delta x)^2}.
\end{aligned}$$ 

The time evolution operator is approximated by:
$$U(\Delta t)=e^{-iH\Delta t} \sim 1-iH\Delta t.$$
Replacing the expression for $H$, we obtain:
$$\psi(x,t+\Delta t)=\psi(x,t)-i[(2\alpha+\Delta t V(x))\psi(x,t)-\alpha(\psi(x+\Delta x,t)+\psi(x-\Delta x,t))],$$ 

with $\alpha=\frac{\Delta t}{(\Delta x)^2}$. The probability of finding an electron at $(x,t)$ is given by
$|\psi(x,t)|^2$. This equations do no conserve this probability exactly, but the error is of the order of $(\Delta t)^2$. The convergence can be
determined by using smaller steps.

We can write this expression explicitly for the real and imaginary parts, becoming:

$$\begin{aligned}
\mathrm{Im} \psi(x, t + \Delta t) = \mathrm{Im} \psi(x, t) + \alpha \mathrm{Re} \psi (x + \Delta x, t) + \alpha \mathrm{Re}\psi(x − \Delta x, t) − (2\alpha + \Delta t V (x)) \mathrm{Re} \psi(x, t) \\
\mathrm{Re} \psi(x, t + \Delta t) = \mathrm{Re} \psi(x, t) −  \alpha \mathrm{Im} \psi (x + \Delta x, t) − \alpha \mathrm{Im} \psi (x − \Delta x, t) + (2\alpha + \Delta tV (x)) \mathrm{Im} \psi(x, t)
\end{aligned}$$


Notice the symmetry between these equations: 
1. the calculation of the imaginary part of the wave function at the later time involves a weighted average of the real part of the wave function at different positions from the earlier time
2.  the calculation of the real part involves a weighted average of the imaginary part for different positions at the earlier time.

This intermixing of the real and imaginary parts of the wave function may seem a bit strange, but remember that this situation is a direct result of our breaking up the wave function into its real and imaginary parts in the first place.

## Stability Conditions

The explicit finite-difference scheme used here is only **conditionally stable**, and the parameter $\alpha = \Delta t / (2\Delta x^2)$ (in our natural units where $\hbar = 1$, $2m = 1$) is what controls it. To see why, substitute a trial solution of the form $\psi_j^n = A^n e^{ikj\Delta x}$ into the update equations — the same Von Neumann analysis used for the heat equation. For the scheme to remain bounded, the amplification factor must satisfy $|A| \leq 1$, which imposes the condition

$$\alpha = \frac{\Delta t}{2\,(\Delta x)^2} \leq \frac{1}{4}$$

in 1D, or more generally $\alpha \leq 1/(4d)$ in $d$ spatial dimensions. This is the direct analog of the stability condition you saw for the heat equation, $\alpha=\Delta t/(\Delta x)^2 \leq 1/(2d)$, and it has the same uncomfortable implication: halving $\Delta x$ forces you to reduce $\Delta t$ by a factor of four to stay stable, making the 2D simulations here quite expensive. In the code, the choice `dt = dx**2 / 20` ensures $\alpha = 1/40$, well within the stable regime for both 1D and 2D. The standard remedy for eliminating this constraint entirely is the [**Crank-Nicolson scheme**](https://en.wikipedia.org/wiki/Crank%E2%80%93Nicolson_method), which averages the spatial operator between the current and next time steps, making the algorithm unconditionally stable at the cost of solving a tridiagonal linear system at each time step — a worthwhile trade-off for production simulations.

# Harmonic Potential 
Let's simulate a Gaussian wave-packet moving along the $x$ axis in a harmonic potential. Recall a harmonic potential is quadratic in nature, $$V(x) = V_{0}x^2$$ where $V_{0}$ is a constant. The harmonic potential is the natural starting point because of it's central importance when approximating potentials around an equilibrium point. Remember around equilibrium, every potential is harmonic! 

The probability distribution for the a particle in the harmonic potential converges to the classical probability as it increases in energy. The simulation below assumes the particle is initial a gaussian packet described by $$\psi(x,0) = A e^{-(x-x_0)^2/2\sigma^2}e^{ikx}.$$
By making this assumption we have a localized, more realistic initial wave packet.

You should compare the results to what you expect from you knowledge of the quantum harmonic oscillator.

<img src="figures/QHM.png" alt="Quantum Harmonic Oscillator" width="400">


```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot
import math
import matplotlib.animation as animation
from IPython.display import HTML


lx = 20
dx = 0.04
nx = int(lx/dx)
dt = dx**2/20.
V0 = 60
alpha = dt/(2*dx**2)

fig = pyplot.figure()
ax = pyplot.axes(xlim=(0, lx), ylim=(0, 2), xlabel='x', ylabel='$|\Psi|^2$')
points, = ax.plot([], [], marker='', linestyle='-', lw=3)

psi0_r = np.zeros(nx+1)
psi0_i = np.zeros(nx+1)

x = np.arange(0, lx+dx, dx)

#Define your potential 
V = np.zeros(nx+1)
V = V0*(x-lx/2)**2

#Initial conditions: wave packet
sigma2 = 0.5**2
k0 = np.pi*10.5 # what happens if k0 = 0?
x0 = lx/2
for ix in range(0,nx):
    psi0_r[ix] = math.exp(-0.5*((ix*dx-x0)**2)/sigma2)*math.cos(k0*ix*dx)
    psi0_i[ix] = math.exp(-0.5*((ix*dx-x0)**2)/sigma2)*math.sin(k0*ix*dx)
    
def solve(i):
    # i made these global to expediate animations, best practices dictate that you should probably pass these functions around 
    global psi0_r, psi0_i

    for ix in range(1,nx-2):
        psi0_i[ix]=psi0_i[ix]+alpha*psi0_r[ix+1]+alpha*psi0_r[ix-1]-(2.*alpha+dt*V[ix])*psi0_r[ix]
        psi0_r[ix]=psi0_r[ix]-alpha*psi0_i[ix+1]-alpha*psi0_i[ix-1]+(2.*alpha+dt*V[ix])*psi0_i[ix]
        
        
    points.set_data(x,psi0_r**2 + psi0_i**2)
    return (points,)

anim = animation.FuncAnimation(fig, solve, frames = 2000, interval=10)

HTML(anim.to_html5_video())            

```

## Single-slit diffraction 
Young’s single-slit experiment consists of a wave passing though a small slit, which causes the emerging wavelets to intefere with each other forming a diffraction pattern. In quantum mechanics, where particles are represented by probabilities, and probabilities by wave packets, it means that the same phenomenon should occur when a particle (electron, neutron) passes though a small slit. Consider a wave packet of some initial width incident on a slit of comparable size. Plotting the probability density $|\psi ^2|$ as the packet crosses the slit, you should see diffraction behavior remiscient of classical wave diffraction patterns. Of course, we need to expand our time evolution equation to 2 dimensions. 

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot
import math

lx      = 20                            #Box length in x
ly      = 20                            #Box length in y
dx      = 0.25                          #Incremental step size in x (Increased this to decrease the time of the sim)
dy      = dx                            #Incremental step size in y
nx      = int(lx/dx)                    #Number of steps in x
ny      = int(ly/dy)                    #Number of steps in y
dt      = dx**2/20.                     #Incremental step size in time
sigma2  = 0.5**2                        #Sigma2 Value
k0      = np.pi*10.5                    #K0 value
amp     = math.pow(1./2., 64)           #Amplitude (to avoid large values out of range. This was one issue)

alpha   = (dt/2.)/dx**2                 #Alpha

psi0_r  = np.zeros(shape=(ny+1,nx+1))   #Initialize real part of psi
psi0_i  = np.zeros(shape=(ny+1,nx+1))   #Initialize imaginary part of psi

#Define your potential wall
V = np.zeros(shape=(ny+1,nx+1))
for ix in range(nx//2-20,nx//2+20):
    for iy in range(0,ny):
        if(abs(iy*dy-ly/2)>2.5):
            V[iy,ix] = 200.

#Initial conditions: wave packet
x0 = 6.
y0 = ly/2.
for x in range(0,nx):
    for y in range(0,ny):
        psi0_r[y,x] = math.exp(-0.5*((x*dx-x0)**2+(y*dy-y0)**2)/sigma2)*math.cos(k0*x*dx)
        psi0_i[y,x] = math.exp(-0.5*((x*dx-x0)**2+(y*dy-y0)**2)/sigma2)*math.sin(k0*x*dx)
                
x = np.arange(0, lx+dx, dx)
y = np.arange(0, ly+dy, dy)
X, Y = np.meshgrid(x, y)

fig = pyplot.figure()
pyplot.contourf(X,Y,psi0_r**2+psi0_i**2)
pyplot.contour(X,Y,V)

#Function to solve incremental changes in psi
def solve(i):
    #Grab psi lists
    global psi0_r, psi0_i

    #Calculate Imaginary Part in all points except for last 2 (because of indice notation)
    for x in range(1,nx-2):
        for y in range(1,ny-2):
            psi0_i[y,x] = psi0_i[y,x] + alpha*psi0_r[y,x+1] + alpha*psi0_r[y,x-1] - (4*alpha + dt*V[y,x])*psi0_r[y,x] + alpha*psi0_r[y+1,x] + alpha*psi0_r[y-1,x] 

    #Calculate Real Part in all points except for last 2 (because of indice notation)
    for x in range(1,nx-2):
        for y in range(1,ny-2):
            psi0_r[y,x] = psi0_r[y,x] - alpha*psi0_i[y,x+1] - alpha*psi0_i[y,x-1] + (4*alpha + dt*V[y,x])*psi0_i[y,x] - alpha*psi0_i[y+1,x] - alpha*psi0_i[y-1,x] 

    

for i in range(1001):
    solve(i)
    if i % 250 == 0: 
        pyplot.contourf(X,Y,psi0_r**2+psi0_i**2);   
        pyplot.contour(X,Y,V);
        pyplot.show()      
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Box dimensions and resolution
lx, ly = 20, 20
dx, dy = 0.25, 0.25
nx, ny = int(lx/dx) + 1, int(ly/dy) + 1  # Ensure proper array sizes
dt = dx**2 / 20.
sigma2 = 0.5**2
k0 = np.pi * 10.5
amp = math.pow(1./2., 64)
alpha = (dt/2.) / dx**2

# Initialize wavefunction arrays and potential
psi0_r = np.zeros((ny, nx))
psi0_i = np.zeros((ny, nx))
V = np.zeros((ny, nx))

# Define potential wall
for ix in range(nx//2-20, nx//2+20):
    for iy in range(0, ny):
        if abs(iy * dy - ly / 2) > 2.5:
            V[iy, ix] = 200.

# Initial wave packet
x0, y0 = 6., ly / 2.
for x in range(nx):
    for y in range(ny):
        psi0_r[y, x] = np.exp(-0.5 * ((x * dx - x0) ** 2 + (y * dy - y0) ** 2) / sigma2) * np.cos(k0 * x * dx)
        psi0_i[y, x] = np.exp(-0.5 * ((x * dx - x0) ** 2 + (y * dy - y0) ** 2) / sigma2) * np.sin(k0 * x * dx)

# Create grid for plotting
x = np.linspace(0, lx, nx)
y = np.linspace(0, ly, ny)
X, Y = np.meshgrid(x, y)

# Setup figure
fig, ax = plt.subplots()
ax.set_xticks([])  # Remove x-axis ticks
ax.set_yticks([])  # Remove y-axis ticks
ax.set_xticklabels([])  # Remove x-axis labels
ax.set_yticklabels([])  # Remove y-axis labels
density = psi0_r**2 + psi0_i**2
cmap_plot = ax.imshow(density / np.max(density), extent=[0, lx, 0, ly], origin='lower', cmap='inferno', animated=True, vmin=0, vmax=1)
ax.contour(X, Y, V, colors='white')

# Wavefunction evolution function
def solve(i):
    global psi0_r, psi0_i

    # Evolve imaginary part
    psi0_i[1:-1, 1:-1] += (
        alpha * (psi0_r[1:-1, 2:] + psi0_r[1:-1, :-2] +
                 psi0_r[2:, 1:-1] + psi0_r[:-2, 1:-1]) 
        - (4 * alpha + dt * V[1:-1, 1:-1]) * psi0_r[1:-1, 1:-1]
    )

    # Evolve real part
    psi0_r[1:-1, 1:-1] -= (
        alpha * (psi0_i[1:-1, 2:] + psi0_i[1:-1, :-2] +
                 psi0_i[2:, 1:-1] + psi0_i[:-2, 1:-1]) 
        - (4 * alpha + dt * V[1:-1, 1:-1]) * psi0_i[1:-1, 1:-1]
    )

    # Compute probability density and normalize
    density = psi0_r**2 + psi0_i**2
    density /= np.max(density)  # Normalize to keep the highest intensity visible

    # Update plot
    cmap_plot.set_array(density)
    return [cmap_plot]

# Create animation
anim = animation.FuncAnimation(fig, solve, frames=1000, interval=10, blit=True)

HTML(anim.to_html5_video())
```

## Double Slit Interference
Young's double-slit experiment is one of the most profound demonstrations of wave-particle duality in quantum mechanics. When a single particle passes through two slits simultaneously, its wavefunction diffracts at each opening and the two emerging wavelets interfere with one another — producing a pattern of alternating high and low probability on the far side. This is not a statistical artifact of many particles: even a single electron, sent through one at a time, builds up the interference pattern over repeated trials. To simulate this, we extend our 2D time-evolution scheme to a potential barrier with *two* slits, and watch the probability density $|\Psi(x,t)|^2$ develop the characteristic constructive and destructive interference fringes.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot
import math
from matplotlib import ticker

lx      = 20                            #Box length in x
ly      = 20                            #Box length in y
dx      = 0.25                          #Incremental step size in x (Increased this to decrease the time of the sim)
dy      = dx                            #Incremental step size in y
nx      = int(lx/dx)                    #Number of steps in x
ny      = int(ly/dy)                    #Number of steps in y
dt      = dx**2/20.                     #Incremental step size in time
sigma2  = 2.0**2                        #Sigma2 Value
k0      = np.pi*10.5                    #K0 value
amp     = math.pow(1./2., 64)           #Amplitude (to avoid large values out of range. This was one issue)

alpha   = (dt/2.)/dx**2                 #Alpha

psi0_r  = np.zeros(shape=(ny+1,nx+1))   #Initialize real part of psi
psi0_i  = np.zeros(shape=(ny+1,nx+1))   #Initialize imaginary part of psi
V       = np.zeros(shape=(ny+1,nx+1))   #Initialize Potential

#Define your potential wall
V = np.zeros(shape=(ny+1,nx+1))
for ix in range(nx//2-20,nx//2+20):
    for iy in range(0,ny):
        if(abs(iy*dy-ly/2)>3.5 or abs(iy*dy-ly/2)<1.5):
            V[iy,ix] = 200.

#Initial conditions: wave packet
x0 = 3.
y0 = ly/2.
for x in range(0,nx):
    for y in range(0,ny):
        psi0_r[y,x] = math.exp(-0.5*((x*dx-x0)**2+(y*dy-y0)**2)/sigma2)*math.cos(k0*x*dx)
        psi0_i[y,x] = math.exp(-0.5*((x*dx-x0)**2+(y*dy-y0)**2)/sigma2)*math.sin(k0*x*dx)
                
x = np.arange(0, lx+dx, dx)
y = np.arange(0, ly+dy, dy)
X, Y = np.meshgrid(x, y)

pyplot.contourf(X,Y,psi0_r**2+psi0_i**2)
pyplot.contour(X,Y,V)

#Function to solve incremental changes in psi
def solve():
    #Grab psi lists
    global psi0_r, psi0_i

    #Calculate Imaginary Part in all points except for last 2 (because of indice notation)
    for x in range(1,nx-2):
        for y in range(1,ny-2):
            psi0_i[y,x] = psi0_i[y,x] + alpha*psi0_r[y,x+1] + alpha*psi0_r[y,x-1] - (4*alpha + dt*V[y,x])*psi0_r[y,x] + alpha*psi0_r[y+1,x] + alpha*psi0_r[y-1,x] 
    #Calculate Real Part in all points except for last 2 (because of indice notation)
    for x in range(1,nx-2):
        for y in range(1,ny-2):
            psi0_r[y,x] = psi0_r[y,x] - alpha*psi0_i[y,x+1] - alpha*psi0_i[y,x-1] + (4*alpha + dt*V[y,x])*psi0_i[y,x] - alpha*psi0_i[y+1,x] - alpha*psi0_i[y-1,x] 
    

for i in range(501):
    solve()
    if i % 100 == 0: 
        prob = psi0_r**2+psi0_i**2
        pyplot.contourf(X,Y,prob);   
        pyplot.contour(X,Y,V);
        pyplot.show()
        pyplot.plot(y,prob[:,-4])
        pyplot.show()

# anim = animation.FuncAnimation(fig, solve, frames = 1000, interval=10)

# HTML(anim.to_jshtml())       
```

*Note:* we see the development of peaks, and also the interference pattern. 

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Box dimensions and resolution
lx, ly = 20, 20
dx, dy = 0.25, 0.25
nx, ny = int(lx/dx) + 1, int(ly/dy) + 1  # Ensure proper array sizes
dt = dx**2 / 20.
sigma2 = 0.5**2
k0 = np.pi * 10.5
amp = math.pow(1./2., 64)
alpha = (dt/2.) / dx**2

# Initialize wavefunction arrays and potential
psi0_r = np.zeros((ny, nx))
psi0_i = np.zeros((ny, nx))
V = np.zeros((ny, nx))

# Define double-slit potential wall
slit_width = 2.0  # Width of each slit
slit_spacing = 4.0  # Space between slits
wall_x = nx // 2  # Position of the wall
barrier_thickness_y = 2  # Reduced thickness in the y-direction

# Create the potential barrier with two slits (Reduced y-thickness)
for ix in range(wall_x - 2, wall_x + 2):  # Keep thin in x-direction (perpendicular to barrier)
    for iy in range(ny):
        y_pos = iy * dy
        if not ((abs(y_pos - ly / 2 - slit_spacing / 2) < slit_width / 2) or
                (abs(y_pos - ly / 2 + slit_spacing / 2) < slit_width / 2)):  
            V[iy, ix] = 200.  # High potential wall, except at slits

# Initial wave packet
x0, y0 = 5., ly / 2.  # Adjusted initial position for better propagation
for x in range(nx):
    for y in range(ny):
        psi0_r[y, x] = np.exp(-0.5 * ((x * dx - x0) ** 2 + (y * dy - y0) ** 2) / sigma2) * np.cos(k0 * x * dx)
        psi0_i[y, x] = np.exp(-0.5 * ((x * dx - x0) ** 2 + (y * dy - y0) ** 2) / sigma2) * np.sin(k0 * x * dx)

# Create grid for plotting
x = np.linspace(0, lx, nx)
y = np.linspace(0, ly, ny)
X, Y = np.meshgrid(x, y)

# Setup figure
fig, ax = plt.subplots()
ax.set_xticks([])  # Remove x-axis ticks
ax.set_yticks([])  # Remove y-axis ticks
ax.set_xticklabels([])  # Remove x-axis labels
ax.set_yticklabels([])  # Remove y-axis labels
density = psi0_r**2 + psi0_i**2
cmap_plot = ax.imshow(density / np.max(density), extent=[0, lx, 0, ly], origin='lower', cmap='inferno', animated=True, vmin=0, vmax=1)
ax.contour(X, Y, V, colors='white')  # Show potential wall/slits

# Wavefunction evolution function
def solve(i):
    global psi0_r, psi0_i

    # Evolve imaginary part
    psi0_i[1:-1, 1:-1] += (
        alpha * (psi0_r[1:-1, 2:] + psi0_r[1:-1, :-2] +
                 psi0_r[2:, 1:-1] + psi0_r[:-2, 1:-1]) 
        - (4 * alpha + dt * V[1:-1, 1:-1]) * psi0_r[1:-1, 1:-1]
    )

    # Evolve real part
    psi0_r[1:-1, 1:-1] -= (
        alpha * (psi0_i[1:-1, 2:] + psi0_i[1:-1, :-2] +
                 psi0_i[2:, 1:-1] + psi0_i[:-2, 1:-1]) 
        - (4 * alpha + dt * V[1:-1, 1:-1]) * psi0_i[1:-1, 1:-1]
    )

    # Compute probability density and normalize
    density = psi0_r**2 + psi0_i**2
    density /= np.max(density)  # Normalize to keep intensity stable

    # Update plot
    cmap_plot.set_array(density)
    return [cmap_plot]

# Create animation
anim = animation.FuncAnimation(fig, solve, frames=1200, interval=10, blit=True)

HTML(anim.to_html5_video())
```
